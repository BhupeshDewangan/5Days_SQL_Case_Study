```python
df['Last Updated']=  pd.to_datetime(df['Last Updated'],errors='coerce')
```

```python
#CLEANING INSTALL COLUMNS
df['Installs']= df['Installs'].str.replace('+','')
df['Installs']= df['Installs'].str.replace(',','')
```

```python
df = df[df.Installs != 'Free']
```

```python
df['Installs']=df['Installs'].astype(int)
# df.Installs.unique()
```

```python
df['Price']=df['Price'].str.replace('$','')
df['Price']= df['Price'].astype(float)
```

```python
df.isnull().sum()
```

```python
avg_rating_by_genre = df.groupby('Genres')['Rating'].mean()

# Fill null values with the average rating of the corresponding genre
df['Rating'].fillna(df['Genres'].map(avg_rating_by_genre), inplace=True)
```
